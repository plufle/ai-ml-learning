Norminal Encoding

```python
import pandas as pd

from sklearn.preprocessing import OneHotEncoder
 
df = pd.DataFrame({
    "color": ["red", "blue", "green", "blue", "red"],
    "size": ["S", "M", "L", "M", "S"],
    "price": [10, 20, 30, 20, 10]
})


encoder = OneHotEncoder()

encoded = encoder.fit_transform(df[["color"]]).toarray()

```

```python
encoded_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out())

encoded_df

df = pd.concat([df, encoded_df], axis=1)
df
```

Label Encoding

```python
df = pd.DataFrame({
    "color": ["red", "blue", "green", "blue", "red"],
    "size": ["S", "M", "L", "M", "S"],
    "price": [10, 20, 30, 20, 10]
})
df
```

```python
from sklearn.preprocessing import LabelEncoder

lbl_encoder = LabelEncoder()
lbl_encoded = lbl_encoder.fit_transform(df['color'])
```

```python
lbl_df = pd.DataFrame(lbl_encoded, columns=["color_encoded"])
lbl_df = pd.concat([df, lbl_df], axis=1)
lbl_df

```

Ordinal Encoding

```python
df = pd.DataFrame({
    "size": ["S", "M", "L", "M", "S"],
})

from sklearn.preprocessing import OrdinalEncoder

oe = OrdinalEncoder(categories=[['S', 'M', 'L']])



oe_df = pd.DataFrame(oe.fit_transform(df), columns=df.columns)

df = pd.concat([df, oe_df], axis=1)
df
```

Target Guided Ordinal Encoding

```python

df = pd.DataFrame({
    "city": ["New York", "Los Angeles", "New York", "Los Angeles", "New York"],
    "price": [10, 20, 30, 20, 10]
})
df
```

```python
mean_city_price = df.groupby("city")["price"].mean().to_dict()
mean_city_price
```

```python
df["city_encoded"] = df["city"].map(mean_city_price)
df
```

```python
import seaborn as sns

data = sns.load_dataset("tips")
data.head()
```

```python
encoding_dict = data.groupby("time")["total_bill"].mean().to_dict()
encoding_dict
```

```python
data["time"] = data["time"].map(encoding_dict)

data
```
