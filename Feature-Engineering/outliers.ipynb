```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = np.array([45,32,56,75,89,54,32,89,90,87,67,54,45,98,99,67,74])

data
```

```python
min_,q1,median,q3,max_ = np.quantile(data, [0,0.25, 0.5, 0.75,1])
min_,q1,median,q3,max_
```

```python
iqr = q3-q1

lower_bound = q1 - 1.5*iqr
upper_bound = q3 + 1.5*iqr

print(lower_bound, upper_bound)
```
