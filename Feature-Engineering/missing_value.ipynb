Missing completely in random (MCAR)

The probability that a value is missing is independent of both observed and unobserved data.
The missingness has nothing to do with the data itself.
It’s essentially pure randomness.

Missing at random (MAR)

The probability that a value is missing is related to observed data, but not the missing value itself.

Missing not at random (MNAR)

The probability that a value is missing is related to the missing value itself, even after considering observed data.

```python
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
# Deleting rows with missing values
df.dropna().shape

# Deleting columns with missing values
df.dropna(axis=1).shape
```

## Impuation Missing Values

```python
# Mean imputation = normal distribution

df["age_mean"] = df["age"].fillna(df["age"].mean())
df[["age","age_mean"]]
 
```

```python
# Median imputation = Outlier based imputation

df["age_median"] = df["age"].fillna(df["age"].median())

df[["age_median","age_mean","age"]]
```

```python
## Mode Imputation

df["embarked_mode"] = df["embarked"].fillna(df["embarked"].mode()[0])
df[["embarked_mode","embarked"]].isnull().sum()
```
