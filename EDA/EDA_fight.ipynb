```python
import pandas as pd

data = pd.read_excel('flight_price.xlsx')
data.info()
```

```python

data["Date"] = data["Date_of_Journey"].str.split("/").str[0]
data["Month"] = data["Date_of_Journey"].str.split("/").str[1]
data["Year"] = data["Date_of_Journey"].str.split("/").str[2]

data["Date"] = data["Date"].astype(int)
data["Month"] = data["Month"].astype(int)
data["Year"] = data["Year"].astype(int)

data.drop("Date_of_Journey", axis=1, inplace=True)
data.head()
```

```python
data["hour_arrival"] = data["Arrival_Time"].str.split(" ").str[0].str.split(":").str[0]
data["hour_arrival"] = data["hour_arrival"].astype(int)
data["minutes_arrival"] = data["Arrival_Time"].str.split(" ").str[0].str.split(":").str[1]
data["minutes_arrival"] = data["minutes_arrival"].astype(int)
data.drop("Arrival_Time", axis=1, inplace=True)
data.head()
```

```python
data["dep_Hour"] = data["Dep_Time"].str.split(":").str[0].astype(int)
data["dep_Minute"] = data["Dep_Time"].str.split(":").str[1].astype(int)
data.drop("Dep_Time", axis=1, inplace=True)
data.head()
```

```python
import numpy as np

data["Total_Stops"] = data["Total_Stops"].map({"non-stop": 0, "1 stop": 1, "2 stops": 2, "3 stops": 3, "4 stops": 4, np.nan: 1})
```

```python
data
```

```python
data.drop("Route", axis=1, inplace=True)
```

```python
data
```

```python
data["Additional_Info"].unique()
```

```python
data["Duration_hour"] = data["Duration"].str.split(" ").str[0].str.split("h").str[0]
data["Duration_hour"] = data["Duration_hour"].replace({"5m": '0.833'})
data["Duration_hour"] = data["Duration_hour"].astype(float)
data.info()
```

```python
data["Duration_minutes"] = data["Duration"].str.split(" ").str[1].str.split("m").str[0]
data["Duration_minutes"] = data["Duration_minutes"].fillna(0)
data.drop("Duration",axis=1,inplace=True)
data
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder()

encoded = encoder.fit_transform(data[["Airline","Source","Destination"]]).toarray()
df_encoded = pd.DataFrame(encoded, columns=encoder.get_feature_names_out(["Airline","Source","Destination"]))
data = pd.concat([data,df_encoded],axis=1)
data.drop(["Airline","Source","Destination"],axis=1,inplace=True)
data
```

```python
data["Additional_Info"].unique()
```

```python
from sklearn.preprocessing import LabelEncoder

oe = LabelEncoder();
lb = oe.fit_transform(data["Additional_Info"])
lb_df = pd.DataFrame(lb, columns=["Additional_Info"])

data = data.drop("Additional_Info", axis=1)
data = pd.concat([data, lb_df], axis=1)
```
