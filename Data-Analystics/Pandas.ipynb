```python
## Series

series = pd.Series([1,2,3,4,5])
print(series)

dict = {"A":1,"B":2,"C":3,"D":4,"E":5}
s = pd.Series(dict)
print(s)

data = [10,20,30 ]
index = ["1","2","3"]
s = pd.Series(data,index)
print(s)
```
